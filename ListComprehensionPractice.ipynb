Read the temperature data from <font face='courier'>NorfolkWeather1999.csv</font> into a variable <font face='courier'>nw1999</font> using list comprehension, using <font face='courier'>readlines()</font> to obtain the data.  Make sure the data is formatted in an appropriate numerical data type.  (Ignore the day index data.)

```python
with open("NorfolkWeather1999.csv",'r') as f:
    nw1999 = f.readlines()
for i in range(len(nw1999)):
    nw1999[i] = nw1999[i].strip().split(',')
    nw1999[i][0] = int(nw1999[i][0])
    nw1999[i][1] = float(nw1999[i][1])
print(nw1999[:5])
ls = []
for i in range(len(nw1999)):
    ls.append(float(nw1999[i][1]))
ls[:5]
```

Read the maxtrix data from <font face='courier'>transMat.txt</font> into a variable <font face='courier'>P</font> using list comprehension.  Try one version using <font face='courier'>readlines()</font> and one that reads the data with <font face='courier'>readline()</font> within the list comprehension.  In the latter approach you will need to use the fact that the data file has 6 rows in it.  Make sure the data is formatted in an appropriate numerical data type.

```python
with open("transMat.txt",'r') as f:
    f = f.readlines()
P = [[float(x) for x in f[i].strip().split(',')] for i in range(6)]
print(P)
```
